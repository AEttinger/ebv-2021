```python
def generate_input_file(dirs_lookup, files_lookup):
    """Generates an input file to be used with RACE Cell segmentation (Stegmaier et al., 2016).
    
    Parameters:
    ----------
    dirs_lookup : dict
        Dictionary containing input and output paths.
    files_lookup : dict
        Dictionary containing file names.
    
    Returns:
    -------
    Nothing. Saves input text file to disk.
    """
    
    input_string = ['--output '+dirs_lookup['output']+'\n',
                '--input 0, '+os.path.join(dirs_lookup['membrane_input'], files_lookup['membrane_file'])+', 2, int\n',
                '--input 1, '+os.path.join(dirs_lookup['csv_input'], files_lookup['csv_file'])+', csv\n',
                '--xml '+os.path.join(dirs_lookup['pipeline'],files_lookup['pipeline_file'])+'\n',
                '--seed 0\n',
                '--lockfile off\n',
                '--subfolder filterid, filtername\n',
                '--outputformat imagename, filtername\n',
                '--end']
    with open(os.path.join(dirs_lookup['input'],files_lookup['input_file']),'w') as f:
        for input_line in input_string:
            f.writelines(input_line)

def run_race(dirs_lookup, files_lookup):
    """Runs XPIWIT with input parameters from current directory."""
    # os.system('./XPIWIT.sh < '+os.path.join(dirs_lookup['input'],files_lookup['input_file']))
    f_input = open(os.path.join(dirs_lookup['input'],files_lookup['input_file']))
    try:
        p = subprocess.run('/home/ies/andreas.ettinger/RACE_Ubuntu_x64/Bin/XPIWIT.sh', stdin=f_input, capture_output=True)
    except Exception:
        error = p.CompletedProcess.stderr
    f_input.close()

    
def save_control_image(dirs_lookup, files_lookup, nuclei, membrane, labels, segmented_membranes):
    fig, ax = plt.subplots(2,2, figsize=(20,20))
    ax[0,0].imshow(nuclei, cmap='gray', vmin = nuclei.min(), vmax=nuclei.max()-0.6*(nuclei.max()-nuclei.min()))
    ax[0,0].axis('off')
    ax[0,1].imshow(nuclei, cmap='gray', vmin = nuclei.min(), vmax=nuclei.max()-0.6*(nuclei.max()-nuclei.min()))
    ax[0,1].imshow(labels, alpha=0.3)
    ax[0,1].axis('off')
    ax[1,0].imshow(membrane, cmap='gray', vmin = nuclei.min(), vmax=nuclei.max()-0.6*(nuclei.max()-nuclei.min()))
    ax[1,0].axis('off')
    ax[1,1].imshow(membrane, cmap='gray', vmin = nuclei.min(), vmax=nuclei.max()-0.6*(nuclei.max()-nuclei.min()))
    ax[1,1].imshow(segmented_membranes, alpha=0.4, cmap=lbl_cmap)
    ax[1,1].axis('off')
    plt.tight_layout()
    plt.savefig(os.path.join(dirs_lookup['control_images_race'], files_lookup['control_file_race']))
    plt.close()
```

```python
import subprocess
```

```python
files_lookup = dict()
files_lookup['pipeline_file'] = 'myRACECSV-param-tune-4.xml'
for label_image in stardist_labels:
    # add file names to dictionary
    basename = re.search('(.+)_Ch4.tiff',os.path.basename(label_image))[1]
    print('processing ...{}'.format(basename))
    files_lookup['nuclei_file'] = basename + '_Ch0.tiff'
    files_lookup['bzlf1_file'] = basename + 'Ch1.tiff'
    files_lookup['membrane_file'] = basename + '_Ch2.tiff'
    files_lookup['gp350_file'] = basename + '_Ch3.tiff'
    files_lookup['csv_file'] = basename + '.csv'
    files_lookup['input_file'] = basename + '.txt'
    files_lookup['membrane_watershed_file'] = basename + '_Ch2_NScaleMorphologicalWatershedFilter_Out1.tif'
    files_lookup['control_file_race'] = basename + '-segmentation-overlay.png'
    
    nuclei = imread(os.path.join(dirs_lookup['nuclei_input'], files_lookup['nuclei_file']))
    membrane = imread(os.path.join(dirs_lookup['membrane_input'], files_lookup['membrane_file']))
    
    labels = imread(label_image)
    sd_props = regionprops(labels)
    # seeds = [ [item.label, round(item.centroid[0]), round(item.centroid[1])] for item in sd_props if item.area > 500 ]
    # try flipped x and y coordinates
    # seeds = [ [item.label, round(item.centroid[1]), round(item.centroid[0])] for item in sd_props if item.area > 500 ]
    # try seed file with xyz
    # seeds = [ [item.label, int(item.centroid[0]), int(item.centroid[1]), 0] for item in sd_props if item.area > 500 ]
    # try without labels
    seeds = [ [item.label, int(item.centroid[1]), int(item.centroid[0]), 0] for item in sd_props if item.area > 500 ]
    
    
    df_seeds = pd.DataFrame(seeds)
    # try with index
    df_seeds.to_csv(os.path.join(dirs_lookup['csv_input'], files_lookup['csv_file']), header=False, sep=';', index=True)
    
    if len(df_seeds) > 0:
        generate_input_file(dirs_lookup, files_lookup)
        run_race(dirs_lookup, files_lookup)
    
    if os.path.exists(os.path.join(dirs_lookup['watershed_output'], files_lookup['membrane_watershed_file'])):
        segmented_membranes = imread(os.path.join(dirs_lookup['watershed_output'], files_lookup['membrane_watershed_file']))  
        save_control_image(dirs_lookup, files_lookup, nuclei, membrane, labels, segmented_membranes)
    print('...done')
```
